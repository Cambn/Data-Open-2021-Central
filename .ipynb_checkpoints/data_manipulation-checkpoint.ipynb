```python
use_us = pd.read_csv('data/tobacco_use_us_cleaned.csv')
```

```python
use_us.head(2)
```

```python
df = use_us.loc[(use_us['QuestionID']=='_RFSMOK3')&(use_us['ResponseID']=='RESP046')&(use_us['Break_Out'].isin(['18-24', '65+']))]
df.head(2)
```

```python
old_vs_young = df.loc[df['Break_Out']=='18-24']\
    .merge(\
           df.loc[df['Break_Out'] == '65+'], 
           how='left', on=['Year','Locationabbr','Locationdesc', 'Topic','Question','Response'])
```

```python
old_vs_young.head()
```

```python
old_vs_young.to_csv('data/old_vs_young.csv', index=False)
```

```python
young_smokers = df = use_us.loc[(use_us['QuestionID']=='_RFSMOK3')&(use_us['ResponseID']=='RESP046')&(use_us['Break_Out'].isin(['18-24']))]
```

```python
COPD_by_state = pd.read_csv('data/disease_by_state.csv').iloc[0:,1:]
```

```python
COPD_by_state
```

```python
disease_vs_smoker = young_smokers.merge(COPD_by_state, how='inner', left_on=['Locationdesc', 'Year'], right_on=['location_name', 'year_id'])
```

```python
disease_vs_smoker.to_csv('data/disease_vs_pct_young_smoker.csv', index=False)
```
